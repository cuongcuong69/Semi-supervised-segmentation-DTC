```python
import pydicom
import os

path = r"D:\NSCLC_Radiomics\NSCLC-Radiomics\LUNG1-001\09-18-2008-StudyID-NA-69331\300.000000-Segmentation-9.554"
files = [f for f in os.listdir(path) if f.lower().endswith(".dcm")]
print("Total files:", len(files))
ds = pydicom.dcmread(os.path.join(path, files[0]))

print("File:", files[0])
print("Modality:", ds.get("Modality", "N/A"))
print("SOPClassUID:", ds.get("SOPClassUID", "N/A"))
print("Manufacturer:", ds.get("Manufacturer", "N/A"))
print("Contains StructureSetROISequence:", hasattr(ds, "StructureSetROISequence"))
print("Contains ROIContourSequence:", hasattr(ds, "ROIContourSequence"))
print("Contains SegmentSequence:", hasattr(ds, "SegmentSequence"))
print("\n---- ROI / Segment Names ----")

if hasattr(ds, "StructureSetROISequence"):
    for roi in ds.StructureSetROISequence:
        print(f"ROI number {roi.ROINumber}: {roi.ROIName}")

if hasattr(ds, "SegmentSequence"):
    for seg in ds.SegmentSequence:
        print(f"Segment number {seg.SegmentNumber}: {seg.SegmentLabel}")
```
